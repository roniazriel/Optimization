```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load csv files

```python
db_6dof = pd.read_csv('6dof_dataset.csv')
db_6dof = db_6dof.drop(columns='Unnamed: 0')
db_6dof.rename(columns = {'Arm ID':'Arm_ID'}, inplace = True)
db_6dof
```

# split arm name to features 

```python

# new data frame with split value columns
new = db_6dof["Arm_ID"].str.split("_", expand=True)

db_6dof["Joint1 type"] = new[1]
db_6dof["Joint1 axis"] = new[2]
db_6dof["Link1 length"] = new[3] + "." + new[4]

db_6dof["Joint2 type"] = new[5]
db_6dof["Joint2 axis"] = new[6]
db_6dof["Link2 length"] = new[7] + "." + new[8]

db_6dof["Joint3 type"] = new[9]
db_6dof["Joint3 axis"] = new[10]
db_6dof["Link3 length"] = new[11] + "." + new[12]

db_6dof["Joint4 type"] = new[13]
db_6dof["Joint4 axis"] = new[14]
db_6dof["Link4 length"] = new[15] + "." + new[16]

db_6dof["Joint5 type"] = new[17]
db_6dof["Joint5 axis"] = new[18]
db_6dof["Link5 length"] = new[19] + "." + new[20]

db_6dof["Joint6 type"] = new[21]
db_6dof["Joint6 axis"] = new[22]
db_6dof["Link6 length"] = new[23] + "." + new[24]

#all_6dof_data = grouped_all_6dof_data.drop(['Arm_ID'], axis=1)

pd.set_option('display.max_columns', 500)
db_6dof
```

```python
'''Indexing Link Configuration and Joint Confuguration'''
import numpy as np
db_6dof['Reached_1'] = np.where(((db_6dof['Point number']== 1) & (db_6dof['Success']== True)), 1, 0)
db_6dof['Reached_2'] = np.where(((db_6dof['Point number']== 2) & (db_6dof['Success']== True)), 1, 0)
db_6dof['Reached_3'] = np.where(((db_6dof['Point number']== 3) & (db_6dof['Success']== True)), 1, 0)
db_6dof['Reached_4'] = np.where(((db_6dof['Point number']== 4) & (db_6dof['Success']== True)), 1, 0)
db_6dof['Reached_5'] = np.where(((db_6dof['Point number']== 5) & (db_6dof['Success']== True)), 1, 0)
db_6dof['Reached_6'] = np.where(((db_6dof['Point number']== 6) & (db_6dof['Success']== True)), 1, 0)
db_6dof['Reached_7'] = np.where(((db_6dof['Point number']== 7) & (db_6dof['Success']== True)), 1, 0)
db_6dof['Reached_8'] = np.where(((db_6dof['Point number']== 8) & (db_6dof['Success']== True)), 1, 0)
db_6dof['Reached_9'] = np.where(((db_6dof['Point number']== 9) & (db_6dof['Success']== True)), 1, 0)
db_6dof['Reached_10'] = np.where(((db_6dof['Point number']== 10) & (db_6dof['Success']== True)), 1, 0)

db_6dof
```

# Group by Manipulator Configurations 

```python
'''Group by Manipulator Configuration
        Success_Rates - how many clusters the arm reached, normalized to 0-1.
        Min_Manipulability - the minimum manipulability value among 10 clusters
        Max_Manipulability - the maximum manipulability value among 10 clusters
        Manipulability_Rates - the avarage of the manipulability index
        MaxSum_Mid_joint_proximity - the maximum sum of mid joint matrix
'''
grouped_all_6dof_data = db_6dof.groupby(['Joint1 type','Joint1 axis', 'Joint2 type', 'Joint2 axis',
                                                      'Joint3 type', 'Joint3 axis', 'Joint4 type',
                                                      'Joint4 axis','Joint5 type', 'Joint5 axis','Joint6 type', 'Joint6 axis',
                                         'Link1 length', 'Link2 length', 'Link3 length', 'Link4 length','Link5 length','Link6 length']).agg(
    Reached_1 = pd.NamedAgg(column="Reached_1", aggfunc='max'),
    Reached_2 = pd.NamedAgg(column="Reached_2", aggfunc='max'),
    Reached_3 = pd.NamedAgg(column="Reached_3", aggfunc='max'),
    Reached_4 = pd.NamedAgg(column="Reached_4", aggfunc='max'),
    Reached_5 = pd.NamedAgg(column="Reached_5", aggfunc='max'),
    Reached_6 = pd.NamedAgg(column="Reached_6", aggfunc='max'),
    Reached_7 = pd.NamedAgg(column="Reached_7", aggfunc='max'),
    Reached_8 = pd.NamedAgg(column="Reached_8", aggfunc='max'),
    Reached_9 = pd.NamedAgg(column="Reached_9", aggfunc='max'),
    Reached_10 = pd.NamedAgg(column="Reached_10", aggfunc='max'),
    Success_Rates=pd.NamedAgg(column="Success", aggfunc='mean'),
    Min_Manipulability=pd.NamedAgg(column='Manipulability - mu', aggfunc='min'),
    Max_Manipulability=pd.NamedAgg(column='Manipulability - mu', aggfunc='max'),
    Manipulability_Rates=pd.NamedAgg(column='Manipulability - mu', aggfunc='mean'),
    MaxSum_Mid_joint_proximity=pd.NamedAgg(column='Sum Mid joint proximity- all joints', aggfunc='max')).reset_index()


grouped_all_6dof_data
```

```python
'''Indexing Link Configuration and Joint Confuguration'''
grouped_all_6dof_data['Joint Configuration Index'] = grouped_all_6dof_data.groupby(['Joint1 type','Joint1 axis', 'Joint2 type', 'Joint2 axis',
                                                      'Joint3 type', 'Joint3 axis', 'Joint4 type',
                                                      'Joint4 axis','Joint5 type', 'Joint5 axis','Joint6 type', 'Joint6 axis']).ngroup()


grouped_all_6dof_data['Link Configuration Index'] = grouped_all_6dof_data.groupby(['Link1 length', 'Link2 length', 'Link3 length', 'Link4 length','Link5 length','Link6 length']).ngroup()

grouped_all_6dof_data
```
