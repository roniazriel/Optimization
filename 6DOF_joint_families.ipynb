```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load csv files

```python
joint_families_6dof = pd.read_csv('6dof_joint_families.csv')
joint_families_6dof = joint_families_6dof.drop(columns='Unnamed: 0')
joint_families_6dof.rename(columns = {'Arm ID':'Arm_ID'}, inplace = True)
joint_families_6dof
```

# split arm name to features 

```python

# new data frame with split value columns
new = joint_families_6dof["Arm_ID"].str.split("_", expand=True)

joint_families_6dof["Joint1 type"] = new[1]
joint_families_6dof["Joint1 axis"] = new[2]
joint_families_6dof["Link1 length"] = new[3] + "." + new[4]

joint_families_6dof["Joint2 type"] = new[5]
joint_families_6dof["Joint2 axis"] = new[6]
joint_families_6dof["Link2 length"] = new[7] + "." + new[8]

joint_families_6dof["Joint3 type"] = new[9]
joint_families_6dof["Joint3 axis"] = new[10]
joint_families_6dof["Link3 length"] = new[11] + "." + new[12]

joint_families_6dof["Joint4 type"] = new[13]
joint_families_6dof["Joint4 axis"] = new[14]
joint_families_6dof["Link4 length"] = new[15] + "." + new[16]

joint_families_6dof["Joint5 type"] = new[17]
joint_families_6dof["Joint5 axis"] = new[18]
joint_families_6dof["Link5 length"] = new[19] + "." + new[20]

joint_families_6dof["Joint6 type"] = new[21]
joint_families_6dof["Joint6 axis"] = new[22]
joint_families_6dof["Link6 length"] = new[23] + "." + new[24]

#all_6dof_data = grouped_all_6dof_data.drop(['Arm_ID'], axis=1)

pd.set_option('display.max_columns', 500)
joint_families_6dof
```

# Group by Manipulator Configurations 

```python
'''Group by Manipulator Configuration
        Success_Rates - how many clusters the arm reached, normalized to 0-1.
        Min_Manipulability - the minimum manipulability value among 10 clusters
        Max_Manipulability - the maximum manipulability value among 10 clusters
        Manipulability_Rates - the avarage of the manipulability index
        MaxSum_Mid_joint_proximity - the maximum sum of mid joint matrix
'''
grouped_all = joint_families_6dof.groupby(['Joint1 type','Joint1 axis', 'Joint2 type', 'Joint2 axis',
                                                      'Joint3 type', 'Joint3 axis', 'Joint4 type',
                                                      'Joint4 axis','Joint5 type', 'Joint5 axis','Joint6 type', 'Joint6 axis','Link1 length', 'Link2 length', 'Link3 length', 'Link4 length','Link5 length','Link6 length']).agg(
    Success_Rates=pd.NamedAgg(column="Success", aggfunc='mean'),
    Min_Manipulability=pd.NamedAgg(column='Manipulability - mu', aggfunc='min'),
    Max_Manipulability=pd.NamedAgg(column='Manipulability - mu', aggfunc='max'),
    Manipulability_Rates=pd.NamedAgg(column='Manipulability - mu', aggfunc='mean'),
    MaxSum_Mid_joint_proximity=pd.NamedAgg(column='Sum Mid joint proximity- all joints', aggfunc='max')).reset_index()


grouped_all
```

```python
'''Indexing Link Configuration and Joint Confuguration'''
grouped_all['Joint Configuration Index'] = grouped_all.groupby(['Joint1 type','Joint1 axis', 'Joint2 type', 'Joint2 axis',
                                                      'Joint3 type', 'Joint3 axis', 'Joint4 type',
                                                      'Joint4 axis','Joint5 type', 'Joint5 axis','Joint6 type', 'Joint6 axis']).ngroup()


grouped_all['Link Configuration Index'] = grouped_all.groupby(['Link1 length', 'Link2 length', 'Link3 length', 'Link4 length','Link5 length','Link6 length']).ngroup()

grouped_all
```

# Group by Joint Configuration

```python
'''Group by Joints Configurations
    Mean_Min_Manipulability - avarage of minimum manipulability
    MaxSum_Mid_joint_proximity - avarage of maximum sum of mid joint matrix
'''

joint_conf_6dof = grouped_all.groupby(['Joint1 type','Joint1 axis', 'Joint2 type', 'Joint2 axis',
                                                      'Joint3 type', 'Joint3 axis', 'Joint4 type',
                                                      'Joint4 axis','Joint5 type', 'Joint5 axis','Joint6 type', 'Joint6 axis']).agg(
    Joint_configuration_index = pd.NamedAgg(column='Joint Configuration Index', aggfunc='mean'),
    Success_Rates = pd.NamedAgg(column='Success_Rates', aggfunc='mean'),
    Mean_Min_Manipulability=pd.NamedAgg(column='Min_Manipulability', aggfunc='mean'),
    Mean_MaxSum_Mid_joint_proximity=pd.NamedAgg(column='MaxSum_Mid_joint_proximity', aggfunc='mean')).reset_index()


joint_conf_6dof['1-Mean_Min_Manipulability'] = 1 - joint_conf_6dof['Mean_Min_Manipulability']
x = joint_conf_6dof['Mean_MaxSum_Mid_joint_proximity']
y = joint_conf_6dof['1-Mean_Min_Manipulability']
h = joint_conf_6dof['Joint_configuration_index']

sns.scatterplot(x=x, y=y, hue=h)
plt.show()

joint_conf_6dof
```
