```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load csv files

```python
arms1 = pd.read_csv('5dof_arms1.csv')
arms2 = pd.read_csv('5dof_arms2.csv')
arms3 = pd.read_csv('5dof_arms3.csv')
arms4 = pd.read_csv('5dof_arms4.csv')
arms5 = pd.read_csv('5dof_arms5.csv')
last_one = pd.read_csv('lastone.csv')
```

```python
all_5dof_data = pd.concat([arms1,arms2,arms3,arms4,arms5,last_one])
all_5dof_data = all_5dof_data.drop(columns='Unnamed: 0')
all_5dof_data.rename(columns = {'Arm ID':'Arm_ID'}, inplace = True)

all_5dof_data
```

# split arm name to features 

```python

# new data frame with split value columns
new = all_5dof_data["Arm_ID"].str.split("_", expand=True)

all_5dof_data["Joint1 type"] = new[1]
all_5dof_data["Joint1 axis"] = new[2]
all_5dof_data["Link1 length"] = new[3] + "." + new[4]

all_5dof_data["Joint2 type"] = new[5]
all_5dof_data["Joint2 axis"] = new[6]
all_5dof_data["Link2 length"] = new[7] + "." + new[8]

all_5dof_data["Joint3 type"] = new[9]
all_5dof_data["Joint3 axis"] = new[10]
all_5dof_data["Link3 length"] = new[11] + "." + new[12]

all_5dof_data["Joint4 type"] = new[13]
all_5dof_data["Joint4 axis"] = new[14]
all_5dof_data["Link4 length"] = new[15] + "." + new[16]

all_5dof_data["Joint5 type"] = new[17]
all_5dof_data["Joint5 axis"] = new[18]
all_5dof_data["Link5 length"] = new[19] + "." + new[20]

#all_6dof_data = grouped_all_6dof_data.drop(['Arm_ID'], axis=1)

pd.set_option('display.max_columns', 500)
all_5dof_data
```

# Check for duplication in DB- same arm was selected more than once

```python
all_5dof_data['counts_arms_duplications'] = all_5dof_data.groupby(['Joint1 type','Joint1 axis', 'Joint2 type', 'Joint2 axis',
                                                      'Joint3 type', 'Joint3 axis', 'Joint4 type',
                                                      'Joint4 axis','Joint5 type', 'Joint5 axis','Link1 length', 'Link2 length', 'Link3 length', 
                                               'Link4 length','Link5 length']).Success.transform(len)
```

```python
testing2 = all_5dof_data.loc[all_5dof_data['counts_arms_duplications'] ==10 ]
testing2
```

# Group by to add problematic arms

```python
grouped_5dof = all_5dof_data.groupby(['Joint1 type','Joint1 axis', 'Joint2 type', 'Joint2 axis',
                                                      'Joint3 type', 'Joint3 axis', 'Joint4 type',
                                                      'Joint4 axis','Joint5 type', 'Joint5 axis','Link1 length', 'Link2 length', 'Link3 length', 'Link4 length','Link5 length']).agg(
    Success_Rates=pd.NamedAgg(column="Success", aggfunc='mean'),
    Min_Manipulability=pd.NamedAgg(column='Manipulability - mu', aggfunc='min'),
    Max_Manipulability=pd.NamedAgg(column='Manipulability - mu', aggfunc='max'),
    Manipulability_Rates=pd.NamedAgg(column='Manipulability - mu', aggfunc='mean'),
    MaxSum_Mid_joint_proximity=pd.NamedAgg(column='Sum Mid joint proximity- all joints', aggfunc='max')).reset_index()

grouped_5dof
```

```python
problematic = pd.read_csv('problematic_5dof.csv')
problematic = problematic.drop(columns='Unnamed: 0')


# new data frame with split value columns
new = problematic["Arm_ID"].str.split("_", expand=True)

problematic["Joint1 type"] = new[1]
problematic["Joint1 axis"] = new[2]
problematic["Link1 length"] = new[3] + "." + new[4]

problematic["Joint2 type"] = new[5]
problematic["Joint2 axis"] = new[6]
problematic["Link2 length"] = new[7] + "." + new[8]

problematic["Joint3 type"] = new[9]
problematic["Joint3 axis"] = new[10]
problematic["Link3 length"] = new[11] + "." + new[12]

problematic["Joint4 type"] = new[13]
problematic["Joint4 axis"] = new[14]
problematic["Link4 length"] = new[15] + "." + new[16]

problematic["Joint5 type"] = new[17]
problematic["Joint5 axis"] = new[18]
problematic["Link5 length"] = new[19] + "." + new[20]

#all_6dof_data = grouped_all_6dof_data.drop(['Arm_ID'], axis=1)

grouped_problematic = problematic.groupby(['Joint1 type','Joint1 axis', 'Joint2 type', 'Joint2 axis',
                                                      'Joint3 type', 'Joint3 axis', 'Joint4 type',
                                                      'Joint4 axis','Joint5 type', 'Joint5 axis','Link1 length', 'Link2 length', 'Link3 length', 'Link4 length','Link5 length']).agg(
    Success_Rates=pd.NamedAgg(column="Success", aggfunc='mean'),
    Min_Manipulability=pd.NamedAgg(column='Manipulability - mu', aggfunc='min'),
    Max_Manipulability=pd.NamedAgg(column='Manipulability - mu', aggfunc='max'),
    Manipulability_Rates=pd.NamedAgg(column='Manipulability - mu', aggfunc='mean'),
    MaxSum_Mid_joint_proximity=pd.NamedAgg(column='Sum Mid joint proximity- all joints', aggfunc='max')).reset_index()

grouped_problematic

all_5 = pd.concat([grouped_5dof,grouped_problematic]).reset_index()
all_5 = all_5.drop(columns='index')
all_5.to_csv('all_5dof_arms.csv')
```
